```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = data.drop(['Timestamp'] ,axis = 1)
```

```python
data
```

```python
data = pd.get_dummies(data).astype(int)
```

```python
data
```

```python
q1 = data["EnergyConsumption"].quantile(0.25)
q3 = data["EnergyConsumption"].quantile(0.75)
iqr = q3-q1
upper_bound = q3+1.5*(iqr)
lower_bound = q1-1.5*(iqr)

data = data[~((data["EnergyConsumption"]>upper_bound))|((data["EnergyConsumption"]<lower_bound))]
```

```python
sn.boxplot(data["EnergyConsumption"])
```

```python
x = data.drop(["EnergyConsumption"],axis = 1)
y = data["EnergyConsumption"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
data = data.fillna(0)
```

```python
data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y , random_state = 42 , test_size = 0.20)
```

```python
x_train.shape , x_test.shape
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
mse = np.mean((y_test-y_pred)**2)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_random = RandomForestRegressor(n_estimators = 150 , max_features = 5)
```

```python
model_random.fit(x_train , y_train)
```

```python
y_pred_rand = model_random.predict(x_test)
```
